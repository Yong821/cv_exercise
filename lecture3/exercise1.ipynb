```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 构建训练样本
x = np.linspace(-5, 5, 200).reshape(100, 2)
x[:, 1] = 1
# 真实的方程为y = 2 * x + 1
y = 2 * x[:, 0].reshape(100, 1) + 1 + np.random.randn(100, 1)
true_y = 2 * x[:, 0].reshape(100, 1) + 1
print(x)
print('-------------------')
print(y)
```

```python
# 构建训练模型
class LinearFit:
    def __init__(self, infeatures):
        self.W = np.zeros(infeatures)
        self.b = 0
        
    def predict(self, x):
        return x @ self.W.T
    
    def evaluate(self, x, y):
        yhat = self.predict(x)
    
    def fit(self, X, Y):
        W_g = np.linalg.inv(X.T @ X) @ X.T @ Y
        self.W = W_g[0]
        self.b = W_g[1]
```

```python
# 训练数据对应的散点图
plt.scatter(x[:, 0], y, s=1, color='red')

# 真实曲线绿色
plt.plot(x[:, 0], true_y, color='green')

cell = LinearFit(2)
cell.fit(x, y)
w, b =  cell.W, cell.b
y1 = w * x[:, 0].reshape(100, 1) + b

# 训练曲线蓝色
plt.plot(x[:, 0], y1, color='blue')
```
